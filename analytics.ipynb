```python
import pandas as pd

df = pd.read_csv('tracks.csv')
print(df.columns)

```

```python
import matplotlib.pyplot as plt
import seaborn as sns

top_genres = df['genre'].value_counts().head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=top_genres.values, y=top_genres.index)
plt.title("Top 10 Genres on Spotify")
plt.xlabel("Number of Tracks")
plt.ylabel("Genre")
plt.tight_layout()
plt.savefig("top_genres.png")
```

```python
top_artists = df['artist_name'].value_counts().head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=top_artists.values, y=top_artists.index)
plt.title("Top 10 Artists by Track Count")
plt.xlabel("Number of Tracks")
plt.ylabel("Artist")
plt.tight_layout()
plt.savefig("top_artists.png")
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df['popularity'], bins=30, kde=True)
plt.title("Track Popularity Distribution")
plt.xlabel("Popularity Score")
plt.tight_layout()
plt.savefig("popularity_distribution.png")
```

```python
features = df[['popularity', 'acousticness', 'danceability', 'energy', 
               'instrumentalness', 'liveness', 'loudness', 
               'speechiness', 'tempo', 'valence']]

plt.figure(figsize=(12,8))
sns.heatmap(features.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Between Audio Features")
plt.tight_layout()
plt.savefig("spotify_feature_correlation.png")
```
